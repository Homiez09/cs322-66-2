```python
import pandas as pd
import numpy as np
```

# 1.เปิดไฟล์ italy_earthquakes_from_2016-08-24_to_2016-11-30.csv| และ อ่านข้อมูลจากไฟล์

```python
time_df = pd.read_csv('data/italy_earthquakes_from_2016-08-24_to_2016-11-30.csv')
time_df.head(5)
```

# 2.ปรับคอลัมน์ Time ให้เป็น Timestamp และตั้งให้เป็น index

```python
time_df = time_df.rename(columns={'Time': 'Timestamp'})
time_df['Timestamp'] = pd.to_datetime(time_df['Timestamp'])
time_df = time_df.set_index('Timestamp')
time_df.head(5)
```

# 3.เวลาที่เกิดแผ่นดินไหวแรงที่สุดคือวันเวลาไหน

```python
time_df['Magnitude'].idxmax()
```

# 4.ในวันที่เกิดแผ่นดินไหวแรงที่สุดมีแผ่นดินไหวกี่ครั้ง

```python
time_df.loc['2016-10-30']["Magnitude"].count()
```

# 5.แสดงข้อมูลตั้งแต่ 7 ถึง 9 โมงเช้าในวันที่เกิดแผ่นดินไหวแรงที่สุด

```python
max_magnitude_date = time_df.sort_index().loc['2016-10-30 07:00:00':'2016-10-30 09:00:00']
max_magnitude_date
```

# 6.แผ่นดินไหวที่แรงที่สุดกับแรงเป็นที่ 2 ห่างกันเท่าไหร่ตอบเป็นวัน-ชั่วโมง-นาที

```python
time_sort_df = time_df.sort_values(['Magnitude'], ascending=False)
time_sort_df.index[0] - time_sort_df.index[1]
```

# 7.คำนวณความแรงของแผ่นดินไหวที่เปลี่ยนไปในแต่ละครั้งของข้อมูลในข้อ 5.หมายเหตุ: ใช้คำสั่ง shift() ช่วยเลื่อนแถวของคอลัมน์ Magnitude เพื่อคำนวณความแตกต่าง

```python
max_magnitude_date['diff'] = max_magnitude_date['Magnitude'].diff()
max_magnitude_date.filter(['Timestamp', 'Magnitude', 'diff'])
```

# 8.มีแผ่นดินไหวเกิดขึ้นกี่ครั้งในแต่ละวัน

```python
time_df['Magnitude'].resample('D').size()
```

# 9.ถ้าแยกเวลาเป็นช่วงช่วงละหนึ่งชั่วโมงมีกี่ช่วงที่ไม่มีแผ่นดินไหวเกิดขึ้นเลยในชั่วโมงนั้น

```python
time_df.resample('H').size()[time_df.resample('H').size() == 0].count()
```

# 10.ถ้าแยกเวลาเป็นช่วงช่วงละ 5 นาทีช่วงเวลาไหนมีค่าเฉลี่ยความรุนแรงของแผ่นดินไหวสูงที่สุดและค่าเฉลี่ยนั้นมีค่าเท่าไหร่

```python
time_df['Magnitude'].resample('5T').mean().max()
```

# 11.ระยะเวลาที่ไม่มีแผ่นดินไหวเลยนานที่สุดคือเท่าไหร่

```python
time_df['time_diff'] = time_df.index.to_series().diff()
time_df['from_time'] = time_df.index.to_series().shift(1)
time_df['time_diff'].max()
```

# 12.ระยะเวลาที่ไม่มีแผ่นดินไหวเลยนานที่สุดคือช่วงไหนระบุเป็นเวลาเร่ิมต้นและเวลาจบ

```python
print("start", time_df[time_df['time_diff'].max() == time_df['time_diff']]['from_time'].iloc[0])
print("end", time_df[time_df['time_diff'].max() == time_df['time_diff']].index[0])
```

# 13.สมมติว่าเรานิยามแผ่นดินไหวที่รุนแรงนั้นต้องมี Magnitude มากกว่าการไหวครั้งก่อนและครั้งหลังอย่างน้อย 2.0 จงคัดเลือกเหตุการณ์แผ่นดินไหวที่รุนแรงออกมา

```python
time_df['Magnitude_diff'] = time_df['Magnitude'].diff()
time_df['Magnitude_after_diff'] = time_df['Magnitude'].shift(-1) - time_df['Magnitude']
strong_eqk = time_df[(time_df['Magnitude_diff'] >= 2) & (time_df['Magnitude_after_diff'] <= -2)].filter(['Latitude', 'Longitude', 'Magnitude'])
strong_eqk
```

# 14.จากตารางข้อ 13. ให้เพิ่มคอลัมน์ใหม่ชื่อ hour_count แสดงจำนวนแผ่นดินไหวในชั่วโมงที่มีเหตุการณ์แผ่นดินไหวรุนแรงด้วย

```python
temp_df = time_df 
temp_df = time_df.resample('H').size()
temp_df.name = 'hour_count'
temp_df = temp_df.to_frame()


strong_eqk = pd.merge_asof(strong_eqk, temp_df, on='Timestamp', direction='backward')
strong_eqk.set_index('Timestamp', inplace=True)
strong_eqk
```

# 15.แสดงจำนวนแผ่นดินไหวแยกตาม 24 ชั่วโมง (เอาจำนวนแผ่นดินไหวของแต่ละชั่วโมง 0-23 ของทุกวันรวมกันแสดงผลตามชั่วโมง)

```python
time_df['Hour'] = time_df.index.hour
time_df.groupby('Hour').size()
```

# 16.คำนวณค่าเฉลี่ยของความรุนแรงแผ่นดินไหวที่มากที่สุดในช่วง 3 วันโดยหาความแรงของแผ่นดินไหวที่มากที่สุดในแต่ละวันก่อนแล้วใช้ sliding window หาค่าเฉลี่ยของทุกๆ 3 วัน

```python
daily_max_magnitude = time_df.resample('D').max()['Magnitude']
avg_max_magnitude = daily_max_magnitude.rolling(3)
for window in avg_max_magnitude:
    print(window)
```
