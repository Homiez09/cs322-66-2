```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../DataSets/tips.csv')
df
```

# การกระจายตัว

```python
sns.displot(data=df, x='tip', kind='hist',
            height=3, aspect=1.5)
```

```python
sns.displot(data=df, x='tip', kind='kde',
            height=3, aspect=1.5)
```

```python
sns.displot(data=df, x='tip', kind='ecdf',
            height=3, aspect=1.5)
```

## แยกหมวดหมู่ข้อมูลด้วยสี

```python
sns.displot(data=df, x='tip', kind='kde', hue='sex',
            height=3, aspect=1.5)
```

```python
sns.displot(data=df, x='tip', kind='hist', hue='time',
            height=3, aspect=1.5)
```

```python
sns.displot(data=df, x='tip', kind='hist', 
            col='time',
            height=3, aspect=1.5)
```

```python
sns.displot(data=df, x='tip', kind='hist', 
            col='time', row='sex',
            height=3, aspect=1.5)
```

## rug

```python
sns.displot(data=df, x='tip', kind='kde', rug=True,
            height=3, aspect=1.5)
```

## hue, row, col

## size, style

```python
sns.relplot(df, x='size', y='total_bill', kind='scatter',
            size='tip',
            height=3, aspect=1.5)
```

```python
sns.relplot(df, y='tip', x='total_bill', kind='scatter',
            style='sex',
            height=3, aspect=1.5)
```

```python
sns.relplot(df, y='tip', x='total_bill', kind='scatter',
            hue='sex', style='day', size='size',
            col='smoker',
            height=3, aspect=1.5)
```

# catplot

```python
df.groupby('day').tip.mean()
```

```python
sns.catplot(df, x='day', y='tip', estimator='mean',
            kind='bar', height=3, aspect=1.5)
```

```python
df['day_cat'] = pd.Categorical(df.day, categories=['Thur','Fri','Sat','Sun'],
               ordered=True)
df
```

```python
sns.catplot(df, x='day_cat', y='tip', estimator='mean',
            kind='bar', height=3, aspect=1.5)
```

```python
sns.catplot(df, x='day_cat', y='tip', estimator='mean',
            kind='point', height=3, aspect=1.5)
```

```python
sns.catplot(df, x='day_cat', y='tip',
            kind='strip', height=3, aspect=1.5)
```

```python
sns.catplot(df, x='day_cat', y='tip',
            kind='swarm', height=3, aspect=1.5)
```

```python
sns.catplot(df, x='day_cat', y='tip',
            kind='box', height=3, aspect=1.5)
```

```python
sns.catplot(df, x='day_cat', y='tip',
            kind='violin', height=3, aspect=1.5)
```

```python
sns.catplot(df, x='day_cat', y='tip',
            hue='sex',
            kind='strip', height=3, aspect=1.5)
```

```python
sns.catplot(df, x='day_cat', y='tip',
            row='sex', col='time', hue='day',
            kind='strip', height=3, aspect=1.5)
```

# pairplot

```python
sns.pairplot(df)
```

# PairGrid

```python
g = sns.PairGrid(df, hue='sex')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```
