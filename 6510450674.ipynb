# 15.คัดเลือก 7 สายการบิน (Operator) ดังนี้ออกมา All Nippon Airways, EVA Air, Scandinavian Airlines (SAS), Singapore Airlines, Swissair, Thai Airways, United Air Lines (ชื่อทั้งหมดอยู่ในไฟล์star_alliance_7.csv) ตั้งชื่อว่า star_7 แล้วแสดงข้อมูลในตารางนี้

# 16.สรุปรวมจำนวนอุบัติเหตุที่เกิดขึ้นของแต่ละสายการบินใน star_7 *ระวัง: เช็คว่ามี 7 สายการบิน

```python
star_7.groupby('Operator')['Operator'].count()
```

# 17.นำอุบัติเหตุของการบินไทย (Thai Airways) มาเรียงลำดับจากล่าสุดย้อนกลับไป

```python
thai_airways = star_7[star_7['Operator'] == 'Thai Airways'].copy()
thai_airways['timestamp'] = pd.to_datetime(thai_airways['Date'])
thai_airways = thai_airways.sort_values('timestamp', ascending=False)
thai_airways
```

# 18.แสดงข้อมูลของ 7 สายการบินเรียงตาม Operator จากน้อยไปมากและในแต่ละ Operator เรียงจากอุบัติเหตุที่เกิดขึ้นก่อนมายังครั้งล่าสุด

```python
star_7['timestamp'] = pd.to_datetime(star_7['Date']).copy()
star_7 = star_7.sort_values(['Operator', 'timestamp'], ascending=[True, True])
star_7
```

# 19.คัดเลือกอุบัติเหตุของสายการบิน Thai Airways ในทศวรรษที่ 90 (ปี 1990–1999) ออกมา

```python
thai_airways_1990s = thai_airways[thai_airways['timestamp'].dt.year.between(1990, 1999)]
thai_airways_1990s
```

# 20.ในตาราง star_7 สกัดคำแรกของคอลัมน์ 'AC Type' ออกมาเป็นชื่อบริษัทที่สร้างเครื่องบินจากนั้นนับจำนวนเครื่องบินที่ตกแยกตามสายการบินและบริษัทเครื่องบิน (ไม่ต้องสนใจข้อมูลที่อาจผิดพลาด)

```python
star_7['Company Name'] = star_7['AC Type'].str.split(' ').str[0]
plane_crash = star_7.groupby(['Operator', 'Company Name']).size().reset_index(name='Crash Count')
plane_crash
```

# 21.นับจำนวนคนตายของ 7 สายการบินแยกตามทศวรรษ (1930, 1940, 1950, ... , 2000) และสายการบิน

```python
fatalities_decade = star_7.copy()
fatalities_decade['Decade'] = (fatalities_decade['timestamp'].dt.year//10)*10
fatalities_decade.groupby(['Decade', 'Operator'])['Fatalities'].sum().reset_index()
```

# 22.เปลี่ยนคอลัมน์ Operator ของ star_7 ให้กลายเป็น category แบบเรียงลำดับโดยลำดับคือลำดับตามตัวอักษร (ตามข้อ 15)

```python
star_7.reset_index()
star_7['Operator'] = pd.Categorical(star_7.Operator, ordered=True,categories=['All Nippon Airways', 'EVA Air','Scandinavian Airlines (SAS)', 'Singapore Airlines', 'Swissair', 'Thai Airways', 'United Air Lines'])
star_7.sort_values('Operator')
```

# 23.แสดงแผนภาพ Histogram และ Rug plot ของ Fatalities Percent ของอุบัตืเหตุจาก 7 สายการบิน

## 23.1 Histogram

```python
sns.displot(data = star_7 , x = 'Fatalities Percent', kind ='hist',height = 3.5, aspect =1)
```

## 23.2 Rug plot

```python
sns.displot(data=star_7, x='Fatalities Percent',kind = 'kde', rug=True, height=3.5, aspect=1)
```

# 24.แสดงกราฟแท่งแสดงจำนวนคนตายรวมของ 7 สายการบิน
