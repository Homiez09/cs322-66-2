# 15.คัดเลือก 7 สายการบิน (Operator) ดังนี้ออกมา All Nippon Airways, EVA Air, Scandinavian Airlines (SAS), Singapore Airlines, Swissair, Thai Airways, United Air Lines (ชื่อทั้งหมดอยู่ในไฟล์star_alliance_7.csv) ตั้งชื่อว่า star_7 แล้วแสดงข้อมูลในตารางนี้

```python
star_alliance_7_df = pd.read_csv('./data/star_alliance_7.csv')
star_alliance_7_df
```

```python
star_7 = air_crash_df[air_crash_df.Operator.isin(star_alliance_7_df['Star Alliance'])]
star_7.head(5)
```

# 16.สรุปรวมจำนวนอุบัติเหตุที่เกิดขึ้นของแต่ละสายการบินใน star_7 *ระวัง: เช็คว่ามี 7 สายการบิน

```python
star_7_accidents = star_7.groupby('Operator').size()
star_7_accidents.name = 'Accident Counts'
```

```python
star_alliance_7_df.join(star_7_accidents, on='Star Alliance').fillna(0)
```

# 19.คัดเลือกอุบัติเหตุของสายการบิน Thai Airways ในทศวรรษที่ 90 (ปี 1990–1999) ออกมา

# 20.ในตาราง star_7 สกัดคำแรกของคอลัมน์ 'AC Type' ออกมาเป็นชื่อบริษัทที่สร้างเครื่องบินจากนั้นนับจำนวนเครื่องบินที่ตกแยกตามสายการบินและบริษัทเครื่องบิน (ไม่ต้องสนใจข้อมูลที่อาจผิดพลาด)

```python
star_7['Company'] = star_7['AC Type'].str.split(' ').str.get(0)
```

```python
star_7.pivot_table(index='Operator', columns='Company', values='Date', aggfunc='count', fill_value=0)
```

# 21.นับจำนวนคนตายของ 7 สายการบินแยกตามทศวรรษ (1930, 1940, 1950, ... , 2000) และสายการบิน

```python
star_7.reset_index(inplace=True)
star_7.head(5)
```

```python
star_7['decade'] = star_7.timestamp.dt.year // 10 * 10
```

```python
star_7.pivot_table(index='Operator', columns='decade', values='timestamp', aggfunc='size', fill_value=0)
```

# 22.เปลี่ยนคอลัมน์ Operator ของ star_7 ให้กลายเป็น category แบบเรียงลำดับโดยลำดับคือลำดับตามตัวอักษร (ตามข้อ 15)

```python
star_7.Operator = pd.Categorical(star_7.Operator, ordered=True)
star_7.sort_values('Operator').head(5)
```

# 23.แสดงแผนภาพ Histogram และ Rug plot ของ Fatalities Percent ของอุบัตืเหตุจาก 7 สายการบิน

```python
# ข้างล่างกับข้างบนคืออันเดียวกันแค่แสดงผลคนละชนิดกัน
sns.displot(data=star_7, x='Fatalities Percent', kind='hist', rug=True)
```

# 24.แสดงกราฟแท่งแสดงจำนวนคนตายรวมของ 7 สายการบิน

```python
sns.catplot(data=star_7, x="Operator", y="Fatalities", estimator="sum", kind="bar", errorbar=None, aspect=2.5)
```

# 25.จากตาราง star_7 แสดงแผนภาพแบบ scatterplot แสดงจำนวนครั้งที่เครื่องบินตกเทียบกับจำนวนคนตายเฉลี่ยต่ออุบัติเหตุหนึ่งครั้งและแยกสีตามสายการบินขนาดของจุดแปรผันตามจำนวนคนตายรวมของสายการบิน

```python
plane_crash_df = star_7.groupby('Operator').Fatalities.agg(['size', 'sum', 'mean'])
plane_crash_df.reset_index(inplace=True) # ใช้ inplace แทนการ assign ค่ากลับไปใหม่
plane_crash_df
```

```python
sns.relplot(data=plane_crash_df, x='size', y='mean', hue='Operator', size='sum')
```
