## hypothesis testing

```python
mean_tip_by_sex = df.groupby('sex').tip.mean()
mean_tip_by_sex
```

H_0: ผู้ชายกับผู้หญิงทิปไม่ต่างกัน ความแตกต่างที่เห็นเป็นความบังเอิญจากการสุ่ม

H_1: ผู้ชายทิปมากกว่าผู้หญิงจริง

```python
alpha = 0.05
```

### resample

```python
real_diff = mean_tip_by_sex['Male'] - mean_tip_by_sex['Female']
real_diff
```

```python
num_by_sex = df.groupby('sex').size()
num_by_sex
```

```python
num_grp_A = num_by_sex['Female']
num_grp_B = num_by_sex['Male']
```

```python
grp_A = df.sample(num_grp_A)
grp_A
```

```python
grp_B = df.loc[ ~df.index.isin( grp_A.index ), :]
grp_B
```

```python
grp_B.tip.mean() - grp_A.tip.mean()
```

```python
ls = []
for i in range(1000):
    grp_A = df.sample(num_grp_A)
    grp_B = df.loc[ ~df.index.isin( grp_A.index ), :]
    mean_tip_diff = grp_B.tip.mean() - grp_A.tip.mean()
    ls.append(mean_tip_diff)
```

```python
ls
```

```python
diff_tip_s = pd.Series(ls)
diff_tip_s
```

```python
(diff_tip_s >= real_diff).sum() / 1000 #p-value
```

p-value สูงกว่า alpha ดังนั้นเราไม่สามารถโต้แย้ง H_0 ได้ เราต้องยอมรับว่าเหตุการณ์ที่เกิดขึ้น (ผู้ชายทิปมากกว่าผู้หญิง) เป็นความบังเอิญ
