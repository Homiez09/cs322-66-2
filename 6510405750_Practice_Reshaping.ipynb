# 8. เติมข้อมูลทวีปที่ยังขาดหายไปในบางประเทศดังนี้ Australasia: OC, Canada/United States/Mexico: NA, Cuba/Jamaica: SA, Czech/Yugoslavia/Germany(All)/Great Britain/Russia(All)/Soviet: EU ประเทศที่เหลือกำหนดให้เป็น 'OT' ไป

```python
continent_data = {
    "OC": ["Australasia"],
    "NA": ["Canada", "United States", "Mexico"],
    "SA": ["Cuba", "Jamaica"],
    "EU": ["Czech", "Yugoslavia", "Germany", "United Team of Germany", "East Germany", "West Germany", "Great Britain", "Russia", "Soviet Union", "Russian Empire"]
}
```

```python
medals["continent"] = medals["continent"].fillna(medals["Country"].apply(lambda x: next((k for k, v in continent_data.items() if x in v), "OT")))
```

```python
medals['continent'].unique()
```

# 16.แสดงตารางรวมเหรียญฤดูร้อนแต่ละประเภทของทุกประเทศในทวีปเดียวกันแสดงผลแยกตามทวีป

```python
summer_medals = long_medals.reset_index()
summer_medals.head(5)
```

```python
summer_medals = summer_medals[summer_medals['Season'] == 'Summer'].pivot_table(index='continent',columns='Medal',values='Count',aggfunc='sum')
summer_medals = summer_medals.rename(columns = {'Gold' : 'SummerGold', 'Silver': 'SummerSilver', 'Bronze': 'SummerBronze'})
summer_medals[['SummerGold','SummerSilver','SummerBronze']]
```

# 17.แสดงตารางสรุปรวมเหรียญแต่ละประเภทของทุกประเทศแยกตามฤดูแข่งขัน

```python
season_medals = long_medals.reset_index()
season_medals.pivot_table(index='Medal',columns='Season',values='Count',aggfunc='sum')
```
